```python
import pandas as pd
```

```python
df = pd.read_csv('bolig_data.csv')
```

```python
df['monthly_aconto'] = df['monthly_aconto'].apply(lambda x: x.replace('kr', '').replace('.', '').replace(',', '').strip() if x else '0').astype(float)
```

```python
df.drop(index=df[df['monthly_aconto']>3131.16].index,inplace=True)
```

```python
df['monthly_aconto'] = df['monthly_aconto'].astype(int).astype(str).apply(lambda x: x+' kr.')
```
